After that we want to start visualizing this data into plotly

```python
fb_rows_to_lists = fb_ad.values.tolist()

phases = ['Cost','Impressions', 'Clicks', 'Leads']
values_c1_g1_ad1 = fb_rows_to_lists[0]
values_c1_g2_ad1 = fb_rows_to_lists[1]
values_c1_g2_ad2 = fb_rows_to_lists[2]
values_c1_g3_ad1 = fb_rows_to_lists[3]
values_c1_g3_ad2 = fb_rows_to_lists[4]
values_c1_g3_ad3 = fb_rows_to_lists[5]
values_c2_g1_ad1 = fb_rows_to_lists[6]
values_c2_g1_ad2 = fb_rows_to_lists[7]
 
colors = ['rgb(32,155,160)', 'rgb(253,93,124)', 'rgb(28,119,139)', 'rgb(182,231,235)', 'rgb(35,154,160)']
```

```python


values = list
n_phase = len(phases)

# the fixed width for the plot
plot_width = 400
 
# height of a section and difference between sections 
section_h = 100
section_d = 10
 
# multiply factor to calculate the width of other sections
unit_width = plot_width / max(values_c1_g1_ad1)
 
# width for all the sections (phases)
phase_w = [int(value * unit_width) for value in values_c1_g3_ad1]

```

```python
height = section_h * n_phase + section_d * (n_phase-1)
 
shapes = []
 
label_y = []
 
for i in range(n_phase):
        if (i == n_phase-1):
                points = [phase_w[i]/2, height, phase_w[i]/2, height - section_h]
        else:
                points = [phase_w[i]/2, height, phase_w[i+1]/2, height - section_h]
 
        path = 'M {0} {1} L {2} {3} L -{2} {3} L -{0} {1} Z'.format(*points)
 
        shape = {
                'type': 'path',
                'path': path,
                'fillcolor': colors[i],
                'line': {
                    'width': 1,
                    'color': colors[i]
                }
    }
        shapes.append(shape)
        
        # Y-axis location for this section's details (phase name and value)
        label_y.append(height - (section_h / 2))
 
        height = height - (section_h + section_d)
```

```python
    # For phase names
    label_trace = go.Scatter(
    x=[-350]*n_phase,
    y=label_y,
    mode='text',
    text=phases,
    textfont=dict(
        color='rgb(150,150,150)',
        size=15
        )
)
 
    # For phase values
    value_trace = go.Scatter(
    x=[350]*n_phase,
    y=label_y,
    mode='text',
    text= values_c1_g3_ad1,
    textfont=dict(
        color='rgb(150,150,150)',
        size=15
        )
)
```
